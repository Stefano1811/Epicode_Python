```python
df_covid_raggruppato = df_covid.groupby(['continent','location'])
```

```python
df_covid_raggruppato = df_covid_raggruppato['total_cases'].max()
```

```python
df_covid_raggruppato
```

```python
df_covid_raggruppato = pd.DataFrame(df_covid_raggruppato).reset_index()
```

```python
df_covid_raggruppato
```

```python
totale_casi = df_covid_raggruppato.groupby('continent')['total_cases'].sum()
```

```python
#NUMERO TOTALE DI CASI PER CONTINENTE
df_totale_casi = pd.DataFrame(totale_casi)
df_totale_casi
```

```python
#Sempre riguardo i casi totali, si chiede di scrivere del codice che, date due variabili contenenti i nomi di continenti,ne confronti i seguenti relativi descrittori statistici:
#valore massimo, media, e percentuale rispetto al numero dei casi totali nel mondo 
#(calcolati anche sulle locazioni senza indicazione di continente).
```

```python
#NUMERO DI CASI NEL MONDO
media_casi_nel_mondo = df_totale_casi.mean()
massimo_di_casi_nel_mondo = df_totale_casi.max()
somma_casi_nel_mondo = df_totale_casi.sum()
```

```python
#NUMERO DI CASI NEL MONDO
print('mean : ',round(media_casi_nel_mondo.values[0]))
print('max : ', massimo_di_casi_nel_mondo.values[0])
print('sum : ',somma_casi_nel_mondo.values[0])
```

```python
continente1 = 'Europe'
continente2 = 'Asia'
```

```python
mms_continenti = df_covid_raggruppato.groupby('continent')['total_cases'].agg(['mean','max','sum'])
```

```python
mms_contagiati_europa = mms_continenti.loc[continente1]
```

```python
#NUMERO CONTAGIATI IN EUROPA
mms_contagiati_europa
```

```python
#NUMERO CONTAGIATI IN ASIA
mms_contagiati_asia = mms_continenti.loc[continente2]
```

```python
# PERCENTUALE DI CONTAGIATI IN EUROPA RISPETTO AL TOTALE DEI CONTAGIATI NEL MONDO 
percentuale_Europa_totale = (mms_contagiati_europa.loc['sum']/somma_casi_nel_mondo) * 100
print('La percentuale di contagiati in Europa rispetto al totale dei contagiati nel mondo è del ',round(percentuale_Europa_totale.values[0],2),'%')
```

```python
# PERCENTUALE DI CONTAGIATI IN ASIA RISPETTO AL TOTALE DEI CONTAGIATI NEL MONDO 
percentuale_Asia_totale = (mms_contagiati_asia.loc['sum']/somma_casi_nel_mondo) * 100
print('La percentuale di contagiati in Asia rispetto al totale dei contagiati nel mondo è del ',round(percentuale_Asia_totale.values[0],2),'%')
```

```python
df_covid['date'] = pd.to_datetime(df_covid['date'])
df_covid['anno'] = df_covid['date'].dt.year
```

```python
filtro_italia = df_covid.location == 'Italy'
```

```python
df_italia = df_covid[filtro_italia]
```

```python
filtro_anno = df_italia.anno == 2022
```

```python
df_ita_2022 = df_italia[filtro_anno]
```

```python
df_ita_2022
```

```python
# EVOLUZIONE DEI CASI TOTALI RISPETTO ALLA DATA
plt.plot(df_ita_2022.date,df_ita_2022.total_cases)
```

```python
#NUMERO DI NUOVI CASI RISPETTO ALLA DATA 
plt.plot(df_ita_2022.date,df_ita_2022.new_cases)
```

```python
# ANDAMENTO DELLA SOMMA CUMULATIVA DEI NUOVI CASI NEL 2022
plt.plot(df_ita_2022.date,df_ita_2022.new_cases.cumsum())
```

```python
# I due grafici (Andamento dei casi totali e andamento della somma cumulativa) sono praticamente uguali perchè il numero dei casi totali deriva dalla somma dei casi giornalieri.
```
